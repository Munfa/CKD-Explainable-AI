# Importing Libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
```

#Dataset-1: CKD_Dataset

```python
df1 = pd.read_excel("/content/CKD_dataset.xlsx")
df1.head()
```

```python
df1 = df1.drop(columns=["p_id"])

df1_X = df1.drop(columns=["d_status"])
df1_y = df1["d_status"]
```

```python
print(df1.duplicated().sum())
```

```python
df1 = df1.drop_duplicates()
```

```python
missing = df1.isnull().sum()

print(missing[missing > 0])
```

```python
print(df1_y.value_counts())
print(df1_y.unique())
```

```python
cat_columns = df1_X.select_dtypes('object').columns

for col in cat_columns:
  df1_X[col] = df1_X[col].astype(str).str.strip().str.lower()
  print(f"{col}: {df1_X[col].unique()}")
```

```python
print(f"d_status: {df1_y.unique()}")
```

```python
oe = OrdinalEncoder()
df1_X = pd.DataFrame(oe.fit_transform(df1_X), columns=df1_X.columns)
```

```python
df1_y = df1_y.map({
    'notckd': 0,
    'ckd': 1
})
```

```python
for col in cat_columns:
  print(f"{col}: {df1_X[col].unique()}")
```

```python
print(f"d_status: {df1_y.unique()}")
```

```python
continuous = [
    "age",
    "ph",
    "sp_g"
]

df1_X[continuous] = df1_X[continuous].apply(pd.to_numeric)
```

```python
df1_clean = pd.concat([df1_X, df1_y], axis=1)
```

```python
df1_clean.to_csv(
    "CKD_preprocessed.csv",
    index=False
)
```

#Dataset-2: Diabetic_CKD_Dataset

```python
df2 = pd.read_csv("/content/DiabeticCKD_dataset.csv")
df2.head()
```

```python
df2.columns
```

```python
df2.columns = df2.columns.str.replace(" ", "_")
```

```python
df2.columns = ['Patient_ID', 'Gender',
       'Job', 'Family_Background_of_Diabetes', 'Height',
       'Diabetic_Year', 'Age', 'Average_Age', 'Weight',
       'Average_Weight', 'BMI', 'Follow_suggested_Diet',
       'Take_Medicine_for_Diabetes',
       'Take_Insulin', 'Hypertension',
       'Heart_Disease', 'Sleep',
       'Water_Consumption',
       'Smoke', 'Zarda_Betel_Leaf',
       'Walk_Regularly', 'Urination_Properly',
       'Urinary_Infection', 'Pain_killer',
       'Calorie_Intake', 'CKD']
```

```python
print(df2.duplicated().sum())
```

```python
df2 = df2.drop_duplicates()
```

```python
groups = df2['Patient_ID']
```

```python
df2 = df2.drop(columns=[
    "Patient_ID",
    "Average_Age",
    "Average_Weight",
    "BMI"
])
```

```python
df2["Current_BMI"] = (
    df2["Weight"] /
    (df2["Height"]/100)**2
)
```

```python
df2 = df2.drop(columns=[
    "Height",
    "Weight"
])
```

```python
missing = df2.isnull().sum()

print(missing[missing > 0])
```

```python
df2.select_dtypes(include='object').columns
```

```python
df2['Gender'].unique()
```

```python
df2['Gender'] = df2['Gender'].map({
    'F': 0,
    'M': 1
})
```

```python
df2.to_csv(
    "DiabeticCKD_preprocessed.csv",
    index=False)
```
